```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
df.head()
```

```python
total_orders = len(df)
avg_delivery_time = df['Delivery_Time_min'].mean()
delay_percentage = df['delay_flag'].mean() * 100

print("Total Orders:", total_orders)
print("Avg Delivery Time:", avg_delivery_time)
print("Delay %:", delay_percentage)
```

```python
delay_time = df.groupby('Time_of_Day')['delay_flag'].mean()

delay_time.plot(kind='bar')
plt.title("Delay Rate by Time of Day")
plt.ylabel("Delay %")
plt.savefig("../images/delay_trend.png")
plt.show()
```

```python
plt.scatter(df['Distance_km'], df['Delivery_Time_min'])
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Time (min)")
plt.title("Distance vs Delivery Time")
plt.savefig("../images/rider_performance.png")
plt.show()
```

```python
plt.figure()
plt.text(0.1, 0.8, f"Total Orders: {total_orders}")
plt.text(0.1, 0.6, f"Avg Delivery Time: {avg_delivery_time:.2f}")
plt.text(0.1, 0.4, f"Delay %: {delay_percentage:.2f}")
plt.axis('off')

plt.savefig("../images/kpi_cards.png")
plt.show()
```
