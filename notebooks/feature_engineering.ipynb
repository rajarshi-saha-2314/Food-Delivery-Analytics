```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned_data.csv")
df.head()
```

```python
df['delay_flag'] = (df['Delivery_Time_min'] > 30).astype(int)
```

```python
def distance_category(x):
    if x < 3:
        return "Short"
    elif x < 7:
        return "Medium"
    else:
        return "Long"

df['distance_category'] = df['Distance_km'].apply(distance_category)
```

```python
df['time_per_km'] = df['Delivery_Time_min'] / df['Distance_km']
```

```python
df['prep_ratio'] = df['Preparation_Time_min'] / df['Delivery_Time_min']
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
```
